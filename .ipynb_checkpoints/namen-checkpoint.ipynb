```python
import spacy
import pandas
import numpy as np
```

```python
namen = pandas.read_csv("../Datenkompetenz_Räumliche_Analyse/Daten/Datenvergleich/saxorum_strassen.tsv", sep = "\t")
```

```python
namen.columns = ["Quelle", "Straßenname", "Wo", "coordinate", "Stadtteil"]
```

```python
namen.head()
```

```python
name_liste = namen["Straßenname"].unique()
```

```python
print (name_liste)
```

die Grundidee:
alle strassennamen von -straße, -weg, -allee usw.
mit spacy alle namen zu erkenn

```python
updated_name_liste = []

for name in name_liste:
    updated_name = name.replace("straße", "").replace("-Straße", "").replace("-Weg", "").replace("weg", "").replace("-", " ").replace("gäßchen", "").replace("Straße", "").replace("Weg", "").replace("Platz", "").replace("platz", "").replace("Am", "").replace("(Langebrück)", "").replace("(Weißig)","").replace("(Cossebaude)", "").replace("(Gohlis)", "")
    updated_name_liste.append(updated_name)

print(len(updated_name_liste))
```

```python
#alle doppelte Namen löchen
updated_name_liste = set(updated_name_liste)
print(len(updated_name_liste))
```

```python
updated_name_liste
```

```python
import spacy

# Spacy-Modell laden (z. B. "de_core_news_sm" für das deutsche Sprachmodell)
nlp = spacy.load("de_core_news_sm")

namen_liste = []

for word in updated_name_liste:
    doc = nlp(word)
    for ent in doc.ents:
        if ent.label_ == "PER":  # Überprüfe, ob die Entität eine Person (Name) ist
            namen_liste.append(ent.text)

print(namen_liste)
```

```python
print (len(namen_liste))
```

```python
namen_liste = sorted(namen_liste)
```

```python
print(namen_liste)
```

```python
namen_liste = list(set(namen_liste))
namen_liste = sorted(namen_liste)
```

```python
print(namen_liste)
```

ich muss ein str erstellen und wieder eine Liste, um doppelte Namen komplett löchen. da es in der liste auf "Anton" und "Anton Graff" gib

```python
namen_str = " ".join(namen_liste)
```

```python
# Teile den String in einzelne Zeichen auf
namen_list = namen_str.split()
```

```python
namen_list = set(namen_list)
namen_list = sorted(namen_list)
```
