```python
data.columns = ["Straßenname", "Personname", "Gender", "Beruf", "coordinate", "PLZ"]
```

```python
# Funktion zum Bereinigen und Formatieren der Koordinaten
def clean_coordinate(coord):
    match = re.search(r'\((.*?)\)', coord)
    if match:
        coord = match.group(1)
    return 'POINT ({})'.format(coord)
```

```python
# Bereinigung und Konvertierung der Koordinaten in die "geometry"-Spalte
data['coordinate'] = data['coordinate'].apply(lambda coord: clean_coordinate(coord))
data['coordinate'] = data['coordinate'].apply(wkt.loads)
```

```python
import pandas as pd
import geopandas as gpd
import folium
from shapely import wkt

# Daten laden und vorbereiten
stadtbezirk = pd.read_csv("/Users/svetlana/Desktop/Strasse_Dresden_GEO/Postleitzahlgebiete.csv", sep=";")
stadtbezirk[["SRID", "geometry"]] = stadtbezirk["geom"].str.split(';', n=1, expand=True)
stadtbezirk['geometry'] = stadtbezirk['geometry'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(stadtbezirk, geometry='geometry')

# CRS festlegen
gdf.crs = "EPSG:4326"  # Beispiel-CRS, bitte anpassen, falls erforderlich

# Gewünschte PLZs
desired_plzs = [1099, 1097, 1069, 1109, 1219, 1309, 1239, 1326, 1067, 1307, 1279, 1217, 1169, 1129, 1259, 1257, 1159, 1157, 1139, 1187, 1156, 1237, 1465, 1445, 1127,1277]

# Filtern der Tabelle basierend auf den gewünschten PLZs
filtered_rows = gdf[gdf['plz'].isin(desired_plzs)]

# GeoDataFrame mit den gefilterten Bezirken erstellen
filtered_gdf = gpd.GeoDataFrame(filtered_rows, geometry='geometry')

# Mittelpunkt der gefilterten Bezirke berechnen
centroid = filtered_gdf.unary_union.centroid


#"#FFB7B7"
# Farben definieren (für jeden eindeutigen PLZ-Wert eine Farbe)
colors = ["#FFF2F2", "#FFE5E6", "#FFDEDD", "#FFDEDD", "#FFD9D9",
               "#FFD3D3", "#FFD3D3", "#FFCECE", "#FFCACA",
               "#FFBAB9", "#FFB7B7", "#FFB7B7", "#FFAAAA",
               "#FF9797", "#FF8D8D", "#FF8787", "#FF8080",
               "#FF8080", "#FF7C7C", "#FF7878", "#FF7D7D",
               "#FF7373", "#FF6D6D", "#FF6B6B", "#FF6262",
               "#FF6060"]

# Karte erstellen
m = folium.Map(location=[centroid.y, centroid.x], zoom_start=11.3)

# Funktion zum Festlegen des Polygonstils basierend auf der PLZ
def style_function(feature):
    plz = feature['properties']['plz']
    color_index = desired_plzs.index(plz) % len(colors)
    plz_color = colors[color_index]
    return {
        'fillColor': plz_color,
        'color': plz_color,
        'weight': 1.5,
        'fillOpacity': 0.7
    }

# Polygone zur Karte hinzufügen und Farben basierend auf der PLZ festlegen
folium.GeoJson(filtered_gdf, style_function=style_function).add_to(m)

# Karte anzeigen
m
```
